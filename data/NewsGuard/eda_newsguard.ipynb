```python
# EDA on the NewsGuard dataset

import pandas as pd
import os

# Load CSV files
myths_df = pd.read_csv("myths-20241219.csv")

# lets filter out the rows where risk of harm is High and language is English
myths_df = myths_df[(myths_df['risk of harm'] == 'High') & (myths_df['Language'] == 'en')]
# only keep the following columns:
myths_df = myths_df[['id', 'name', 'description', 'category']]
myths_df
```

```python
# convert myths_df to jsonl
myths_df.to_json("myths.jsonl", orient='records', lines=True)

```
